```python
import math
import numpy as np

def calculadora():
    # Exibe o menu de opções para o usuário
    print("Escolha a operação:")
    print("1. Adição")
    print("2. Subtração")
    print("3. Multiplicação")
    print("4. Divisão")
    print("5. Seno")
    print("6. Cosseno")
    print("7. Tangente")
    print("8. Probabilidade de um evento simples")
    print("9. Probabilidade conjunta de dois eventos independentes")
    print("10. Adição de Matrizes")
    print("11. Multiplicação de Matrizes")

    # Recebe a escolha do usuário
    escolha = input("Digite sua escolha (1-11): ")

    if escolha in ['1', '2', '3', '4']:
        # Operações aritméticas básicas
        print("Você escolheu uma operação aritmética básica.")
        num1 = float(input("Digite o primeiro número: "))
        num2 = float(input("Digite o segundo número: "))

        if escolha == '1':
            # Adição
            print(f"Adição: {num1} + {num2} = {num1 + num2}")
        elif escolha == '2':
            # Subtração
            print(f"Subtração: {num1} - {num2} = {num1 - num2}")
        elif escolha == '3':
            # Multiplicação
            print(f"Multiplicação: {num1} * {num2} = {num1 * num2}")
        elif escolha == '4':
            # Divisão
            if num2 != 0:
                print(f"Divisão: {num1} / {num2} = {num1 / num2}")
            else:
                print("Erro: Divisão por zero não é permitida")

    elif escolha in ['5', '6', '7']:
        # Operações trigonométricas
        print("Você escolheu uma operação trigonométrica.")
        angulo = float(input("Digite o ângulo em graus: "))
        radianos = math.radians(angulo)

        if escolha == '5':
            # Cálculo do seno
            print(f"Seno de {angulo}° é {math.sin(radianos)}")
        elif escolha == '6':
            # Cálculo do cosseno
            print(f"Cosseno de {angulo}° é {math.cos(radianos)}")
        elif escolha == '7':
            # Cálculo da tangente
            print(f"Tangente de {angulo}° é {math.tan(radianos)}")

    elif escolha == '8':
        # Cálculo da probabilidade de um evento simples
        print("Você escolheu calcular a probabilidade de um evento simples.")
        favoraveis = float(input("Digite o número de resultados favoráveis: "))
        possiveis = float(input("Digite o número total de resultados possíveis: "))
        if possiveis != 0:
            probabilidade = favoraveis / possiveis
            print(f"A probabilidade do evento é {probabilidade}")
        else:
            print("Erro: O número total de resultados possíveis não pode ser zero")

    elif escolha == '9':
        # Cálculo da probabilidade conjunta de dois eventos independentes
        print("Você escolheu calcular a probabilidade conjunta de dois eventos independentes.")
        prob1 = float(input("Digite a probabilidade do primeiro evento (entre 0 e 1): "))
        prob2 = float(input("Digite a probabilidade do segundo evento (entre 0 e 1): "))
        if 0 <= prob1 <= 1 and 0 <= prob2 <= 1:
            prob_conjunta = prob1 * prob2
            print(f"A probabilidade conjunta dos dois eventos é {prob_conjunta}")
        else:
            print("Erro: As probabilidades devem estar entre 0 e 1")

    elif escolha == '10':
        # Adição de Matrizes
        print("Você escolheu a adição de matrizes.")
        linhas = int(input("Digite o número de linhas da matriz: "))
        colunas = int(input("Digite o número de colunas da matriz: "))
        print("Digite os elementos da primeira matriz:")
        matriz1 = []
        for i in range(linhas):
            linha = list(map(float, input().split()))
            matriz1.append(linha)
        print("Digite os elementos da segunda matriz:")
        matriz2 = []
        for i in range(linhas):
            linha = list(map(float, input().split()))
            matriz2.append(linha)
        resultado = np.add(matriz1, matriz2)
        print(f"Resultado da adição de matrizes: {resultado}")

    elif escolha == '11':
        # Multiplicação de Matrizes
        print("Você escolheu a multiplicação de matrizes.")
        linhas1 = int(input("Digite o número de linhas da primeira matriz: "))
        colunas1 = int(input("Digite o número de colunas da primeira matriz: "))
        print("Digite os elementos da primeira matriz:")
        matriz1 = []
        for i in range(linhas1):
            linha = list(map(float, input().split()))
            matriz1.append(linha)
        linhas2 = int(input("Digite o número de linhas da segunda matriz: "))
        colunas2 = int(input("Digite o número de colunas da segunda matriz: "))
        print("Digite os elementos da segunda matriz:")
        matriz2 = []
        for i in range(linhas2):
            linha = list(map(float, input().split()))
            matriz2.append(linha)
        if colunas1 == linhas2:
            resultado = np.dot(matriz1, matriz2)
            print(f"Resultado da multiplicação de matrizes: {resultado}")
        else:
            print("Erro: O número de colunas da primeira matriz deve ser igual ao número de linhas da segunda matriz")

# Chama a função calculadora
calculadora()
```
